```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
df1= pd.read_csv('/workspaces/jeopardy_hypths_1/data/Jeopardy .csv')
print(df1.shape)
print(list(df1.columns))
```

```python
df1.columns= (
    df1.columns
    .str.strip()
    .str.lower()
    .str.replace(' ','_')
)

print(list(df1.columns))

v1= df1[['question','answer','air_date','value']].copy()
v1.head(3)
```

```python
v1['air_date']= pd.to_datetime(v1['air_date'],errors= 'coerce')
v1['year']= v1['air_date'].dt.year
print(v1[['air_date','year']].head(5))
print('rows that have no year:', v1['year'].isna().sum())
```

```python
k_t='/workspaces/jeopardy_hypths_1/data/kids_teen.tsv'
kids= pd.read_table(k_t)
print(kids.shape)
print(list(kids.columns))
kids.head(5)
```

```python
v2= kids[['question','answer','air_date','value']].copy()
v2['year']= pd.to_numeric(v2['air_date'].astype(str).str[:4],errors= 'coerce')
print(v2.shape)
v2.head(5)
print('rows with no year:', v2['year'].isna().sum())
```

```python
m_t='/workspaces/jeopardy_hypths_1/data/master_1-35.tsv'
master= pd.read_table(m_t)
print(master.shape)
print(list(master.columns)[:12])
master.head(3)
```

```python
v3=master[['question','answer','air_date','value']].copy()
v3['year']= pd.to_numeric(v3['air_date'].astype(str).str[:4], errors= 'coerce')
print(v3.shape)
v3.head(5)
print('rows with no year',v3['year'].isna().sum())
```

```python
allt= pd.concat([v1,v2,v3], ignore_index=True)
print('combined shape:', allt.shape)
allt.head(5)
```

```python
allt=allt[allt['year'].notna() & allt['answer'].astype(str).str.strip().ne('')].copy()
print(allt.shape)
```
